```python
#ENTER THE FILENAMES INTO A LIST
FILENAMES = ['result_reproduce_fit0_its100000_addprob1_pref2_time27_02_55_20.csv',\
             'result_reproduce_fit0_its100000_addprob1_pref2_time27_15_36_44.csv',\
             'result_reproduce_fit0_its100000_addprob1_pref2_time27_15_36_51.csv']
```

```python
import pandas as pd
```

```python
# load the data
datalist = []
for filenumber in range(len(FILENAMES)):
    datalist.append(pd.read_csv(FILENAMES[filenumber]))
```

```python
# make the simulation numbers consecutively
max_sim = 0
for dataframe in datalist:
    dataframe['sim'] += max_sim
    max_sim = max(dataframe['sim'])+1

# concatinate the dataframes
merged_dataframes = pd.concat(datalist)
# save the csv
merged_dataframes.to_csv('merged_dataframes_pref2.csv', index=False)
```
